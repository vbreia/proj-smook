```python
# Imports
import seaborn as sea
import random
import numpy as np
import pandas as pd
import matplotlib as mat
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

## Criação do Dataframe

```python
# Valores randômicos
np.random.seed(42)
n = 1000
pct_smokers = 0.2

# Variáveis
flag_fumante = np.random.rand(n) < pct_smokers
idade = np.random.normal(40, 10, n)
altura = np.random.normal(170, 10, n)
peso = np.random.normal(70, 10, n)

# Dataframe
dados = pd.DataFrame({'altura': altura, 'peso': peso, 'flag_fumante': flag_fumante})

# Cria os dados para a variável flag_fumante
dados['flag_fumante'] = dados['flag_fumante'].map({True: 'Fumante', False: 'Não Fumante'})
```

```python
dados.head()
```

## Criação do plot com o Seaborn

```python
# Style
sea.set(style = "ticks")

# lmplot
sea.lmplot(x = 'altura', 
           y = 'peso', 
           data = dados, 
           hue = 'flag_fumante', 
           palette = ['tab:cyan', 'tab:red'], 
           height = 7)

# Labels e título
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.title('Relação Entre Altura e Peso de Fumantes e Não Fumantes')

# Remove as bordas
sea.despine()

# Show
plt.show()
```

### Com o Linear Model Plot é possível analisar detalhadamente se há relação entre os dados (*altura e peso*) e se a pessoa é ou não fumante.
- *Colorido em ciano, temos filtrados os indivíduos não fumantes*
- *Colorido em vermelho, temos filtrados os indvíduos fumantes*

### na linha horizontal, é possível identificar a relação entre os dados informados e o fato dos indivíduos:
- Fumante (*Colorido em vermelho*) - Pelo fato da linha horizontal ter uma inclinação crescente, tem uma significativa influência com o peso e a altura.
- Não fumante (*Colorido em ciano*) - Não tem uma relação significativa com o peso e altura.
